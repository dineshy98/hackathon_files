```python
# function that returns unique skills when we pass skills column



def make_unique(col):
    all_skills = []
    for i in list(col):
        i = str(i)
        for j in i.split(','):
            all_skills.append(j)
    
    temp = []
    for k in all_skills:
        if k == np.nan:
           print(k)
        else:
            temp.append(k)
        all_skills = temp
        
    
    all_skills = [str(j) for j in all_skills]
    all_skills = [j.lower() for j in all_skills]
    all_skills = [str(j) for j in all_skills]
    all_skills = [j.strip() for j in all_skills]
    all_skills = [j.replace('.' , '') for j in all_skills]
    unique_skills = list(set(all_skills))
    
    return unique_skills
```

```python
# concating all unique skills of train,test,project

p = make_unique(project['skills'])
q = make_unique(train['skills']) 
r = make_unique(test['skills'])


unique_skills = list(set(p+q+r))
unique_skills.sort()

unique_skills = pd.DataFrame(unique_skills,columns = ['skill'])
unique_skills.dropna(inplace = True)

unique_skills.head(25)
```

```python
# function that creates dataframe of columns id and skill when we pass id and sring of skills


def make_df_for_row(candidate,skills):
    
    skills = str(skills)
    skills = list(skills.split(','))
    
    
    #removes nan
    temp = []
    for k in skills:
        if k == np.nan:
           print(k)
        else:
            temp.append(k)
        skills = temp
     
    
    
    skills = [str(j) for j in skills]
    skills = [j.lower() for j in skills]
    skills = [str(j) for j in skills]
    skills = [j.strip() for j in skills]
    skills = [j.replace('.' , '') for j in skills]
    
    temp_dict = []
    for skill in skills:
        
        temp_dict.append({
            
            'name': candidate,
            'skill': skill
            
            })
        
    return pd.DataFrame(temp_dict)
```

## 3.Creating DataFrame to replace with dictionary

```python
# Now we have a function for just 1 row

#for one go, I have created another function that creates dataframe for multiple rows using above function

def make_df_multiple_rows(col1,col2):
    
    main_df = pd.DataFrame()
    for i,j in zip(list(col1),list(col2)):
    
        temp_df = make_df_for_row(i,j)
        main_df = pd.concat([main_df,temp_df])
        
    return main_df
```
